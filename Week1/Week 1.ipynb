### Feature selection

```python
# correlation matrix
corr = df.corr("pearson")
corr.style.background_gradient(cmap='coolwarm')
```

```python
# combine some of the features
df["Total_Bathroom"] = (df["Bsmt_Full_Bath"] + (0.5 * df["Bsmt_Half_Bath"])
                       + df["Full_Bath"] + (0.5 * df["Half_Bath"]) )
df["Total_Home_Quality"] = df["Overall_Qual"] + df["Overall_Cond"]
```

```python
# remove unnecessary features that got combined or don't correlate to sale price
df_2 = df.drop(["Bsmt_Full_Bath", "Bsmt_Half_Bath", "Full_Bath",
               "Half_Bath", "Overall_Qual", "Overall_Cond", "Three_season_porch",
                "Screen_Porch", "Pool_Area", "Misc_Val", "Mo_Sold", "Year_Sold"], axis=1)
df_2.head()
```

### Train ml models

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.linear_model import LinearRegression, BayesianRidge
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from mlxtend.regressor import StackingRegressor
```

```python
# train test split
train = df_2[0:2000]
test = df_2[2000:]
print(len(train))
print(len(test))
```

```python
# seperate X and Y
# np.log1p produces only positive values and removes the 'danger' of large negative numbers
# source: https://scicomp.stackexchange.com/questions/20629/when-should-log1p-and-expm1-be-used
X_train = train.drop(["Sale_Price"], axis=1)
Y_train = np.log1p(train["Sale_Price"])

X_test = test.drop(["Sale_Price"], axis=1)
Y_test = np.log1p(test["Sale_Price"])
```

```python
# One hot encoding categorical features since models can't process strings
X_train_dummy = pd.get_dummies(X_train)
X_test_dummy = pd.get_dummies(X_test)
```

```python
# cross fold split to get better result
kf = KFold(n_splits=10, random_state=42, shuffle=True)

# source: https://www.kaggle.com/code/marto24/beginners-prediction-top3
def cv_rmse(model):
    rmse = np.sqrt(-cross_val_score(model, X_train_dummy, Y_train, scoring="neg_mean_squared_error", cv=kf))
    return (rmse)
```

```python
# Linear Regression
lreg = LinearRegression()
# Bayesian Ridge Regression
brr = BayesianRidge(compute_score=True)
# Gradient Boost Regressor
gbr = GradientBoostingRegressor()
# Random Forest Regressor
rfr = RandomForestRegressor()

# Stacked Regressor
stack_gen = StackingRegressor(regressors=(LinearRegression(),
                                          BayesianRidge(),
                                          GradientBoostingRegressor(),
                                          RandomForestRegressor()),
                              meta_regressor = GradientBoostingRegressor(),
                              use_features_in_secondary = True)
```

```python
cv_scores = []

# linear Regression
lreg_score = cv_rmse(lreg)
cv_scores.append(lreg_score.mean())

# Bayesian Ridge Regression
brr_score = cv_rmse(brr)
cv_scores.append(brr_score.mean())

# Gradient Boost Regressor
gbr_score = cv_rmse(gbr)
cv_scores.append(gbr_score.mean())

# Random Forest Regressor
rfr_score = cv_rmse(rfr)
cv_scores.append(rfr_score.mean())

# Stacked Regressor
stack_gen_score = cv_rmse(stack_gen)
cv_scores.append(stack_gen_score.mean())
```

```python
models_name = ["Linear Regression", "Bayesian Rigde Regression", "Gradient Boost Regressor",
         "Random Forest Regressor", "Stacked Regressor"]
d = {'Model': models_name, 'cv_rmse': cv_scores}
result_cv = pd.DataFrame(d)
result_cv
```

Linear Regression is the best model in cross validation, but we haven't utilize Hyperparameter tuning for other models yet

### Hyperparameter tuning

```python
from sklearn.model_selection import RandomizedSearchCV
```

#### Hyperparameter tuning on random_forest

```python
# hyperparameter tuning on random_forest
rfr.get_params()
```

Explaination for rfr params
<br> bootstrap: Bootstrap means that instead of training on all the observations, each tree of RF is trained on a subset of the observations. 
<br> max_features: Number of features to consider at every split
<br> min_samples_split: Minimum number of samples required to split a node
<br> min_samples_leaf: Minimum number of samples required at each leaf node
<br> n_estimators: Number of trees in random forest

source: 
<br>1. https://medium.com/analytics-vidhya/bootstrapping-and-oob-samples-in-random-forests-6e083b6bc341#:~:text=Bootstrap%20means%20that%20instead%20of,all%20the%20trees%20are%20aggregated
<br>2. https://towardsdatascience.com/hyperparameter-tuning-the-random-forest-in-python-using-scikit-learn-28d2aa77dd74

```python
# define hyperparameter values
bootstrap = [True, False]
max_features = ['auto', 'sqrt']
min_samples_split = [2, 5, 10]
min_samples_leaf = [1, 2, 4]
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 300, num = 10)]

random_grid = {
    "bootstrap": bootstrap,
    "max_features": max_features,
    "min_samples_split": min_samples_split,
    "min_samples_leaf": min_samples_leaf,
    "n_estimators": n_estimators
}
```

```python
# n_iter controls the number of different distributions to try
rfr_random = RandomizedSearchCV(estimator = rfr, 
                                param_distributions = random_grid,
                                n_iter = 100,
                                cv = 3,
                                verbose=2,
                                random_state=42,
                                n_jobs = -1)
```

```python
rfr_random.fit(X_train_dummy, Y_train)
```

```python
# best hyperparameters params for rfr
rfr_random.best_params_
```

```python
# rfr with selected hyperparameters
rfr_2 = RandomForestRegressor(n_estimators=188, min_samples_split=5,
                             min_samples_leaf=2, max_features= "auto",
                             bootstrap=False)
```

```python
rfr_2_score = cv_rmse(rfr_2)
```

```python
print("rsme of rfr2:", rfr_2_score.mean())
```

As you can see, the model has better result than linear regression (0.20 > 0.15) and also then before (0.20 > 0.14)

#### Apply model on test dataset

```python
def cv_rmse_test(model):
    rmse = np.sqrt(-cross_val_score(model, X_test_dummy, Y_test, scoring="neg_mean_squared_error", cv=kf))
    return (rmse)
```

```python
# Stacked Regressor
stack_gen_2 = StackingRegressor(regressors=(LinearRegression(),
                                          BayesianRidge(),
                                          GradientBoostingRegressor(),
                                          RandomForestRegressor(n_estimators=188, min_samples_split=5,
                                                                 min_samples_leaf=2, max_features= "auto",
                                                                 bootstrap=False)),
                              meta_regressor = GradientBoostingRegressor(),
                              use_features_in_secondary = True)
```

```python
cv_scores_test = []

# linear Regression
lreg_score_test = cv_rmse_test(lreg)
cv_scores_test.append(lreg_score_test.mean())

# Bayesian Ridge Regression
brr_score_test = cv_rmse_test(brr)
cv_scores_test.append(brr_score_test.mean())

# Gradient Boost Regressor
gbr_score_test = cv_rmse_test(gbr)
cv_scores_test.append(gbr_score_test.mean())

# Random Forest Regressor 2 with hyperparameter tuning
rfr_score_test = cv_rmse_test(rfr_2)
cv_scores_test.append(rfr_score_test.mean())



# Stacked Regressor with rfr 2
stack_gen_score_test = cv_rmse(stack_gen_2)
cv_scores_test.append(stack_gen_score_test.mean())
```

```python
models_name_2 = ["Linear Regression", "Bayesian Rigde Regression", "Gradient Boost Regressor",
         "Random Forest Regressor 2", "Stacked Regressor 2"]
d = {'Model': models_name_2, 'cv_rmse': cv_scores_test}
result_cv = pd.DataFrame(d)
result_cv
```

####  Random Forest Regressor with hyperparameter tuning yields the best result in comparison to other models
